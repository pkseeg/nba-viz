```python
from py_ball.league import League
from py_ball.player import Player

from basketball_reference_scraper.seasons import get_schedule
from basketball_reference_scraper.box_scores import get_box_scores


import pandas as pd
from numpy import std
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

from utils import name_swap, set_size
from config import py_ball_headers

# for notebook purposes
from tqdm.notebook import tqdm, trange
```

```python
def get_fg3_stddev(team, season):
    s = get_schedule(season, playoffs=False)
    t = s[(s.VISITOR == name_swap[team]) | (s.HOME == name_swap[team])]
    player_fg3p = {}
    missing = []
    for r in trange(len(t)):
        d = str(t.iloc[r]['DATE'])[:10]
        home = t.iloc[r]['HOME']
        visitor = t.iloc[r]['VISITOR']
        h = name_swap[home]
        v = name_swap[visitor]
        try:
            b = get_box_scores(d, v, h, period='GAME', stat_type='BASIC')[team]
            # check triple doubles
            for i, row in b.iterrows():
                player = row['PLAYER']
                if player == 'Team Totals': continue
                if player not in player_fg3p: player_fg3p[player] = [] # keep track of this player if we haven't before
                if not type(row['3P%']) == float: # check for nans
                    if row['3P%'] != 'Did Not Play': # man that was an annoying bug
                        val = float(row['3P%'])
                        player_fg3p[player].append(val)     
        except BaseException as e:
            missing.append((d, h, v))
    return player_fg3p
```

```python
l = League(py_ball_headers, season='2021-22', current_season=1)
```

```python
player_ids = [(p['PERSON_ID'],p['DISPLAY_FIRST_LAST']) for p in l.data['CommonAllPlayers']]
team_ids = [p['TEAM_ABBREVIATION'] for p in l.data['CommonAllPlayers']]
print(len(player_ids)) # we have this many player ids
```

```python
team_ids = set(team_ids)
team_ids.remove('')
len(team_ids)
```

```python
len(player_ids) == len(set(player_ids)) # these are unique values, good
```

```python
team_consistency = {}
for team in tqdm(team_ids):
    print(team)
    team_consistency[team] = get_fg3_stddev(team, 2022)
```

```python
player_consistency = {}
for team in team_consistency:
    for player in team_consistency[team]:
        if player not in player_consistency:
            player_consistency[player] = team_consistency[team][player]
        else:
            # check which is longer (whichever team they played for more, that's what they'll count as)
            if len(player_consistency[player]) < len(team_consistency[team][player]):
                player_consistency[player] = team_consistency[team][player]
print(len(player_consistency))
```

```python
names = []
stddev = []
for p in player_consistency:
    names.append(p)
    stddev.append(std(player_consistency[p]))
player_consistency_df = pd.DataFrame({'name': names,
                                      'fg3_stddev': stddev})
```

```python
player_consistency_df.tail()
```

```python
fg3_pct = []
fg3a = []
ids = []
team_ids = []
names = []
for p_id, p_name in tqdm(player_ids):
    p = Player(py_ball_headers, player_id=p_id)
    if len(p.data['SeasonTotalsRegularSeason']) == 0: continue
    s = p.data['SeasonTotalsRegularSeason'][-1]
    p_team = s['TEAM_ABBREVIATION']
    if s['SEASON_ID'] != '2021-22': continue
    fg3_pct.append(s['FG3_PCT'])
    fg3a.append(s['FG3A'])
    #fg3_stddev.append(get_fg3_stddev(p_name, p_team))
    ids.append(p_id)
    team_ids.append(s['TEAM_ID'])
    names.append(p_name)
```

```python
consistency_df = pd.DataFrame({'player_id': ids,
                               'team_id': team_ids,
                               'name': names,
                               'fg3_pct': fg3_pct, 
                               'fg3a': fg3a})
```

```python
consistency_df.head()
```

```python
# merge by name
consistency = pd.merge(consistency_df,player_consistency_df,on='name')
consistency.head()
```

```python
consistency.to_csv('data/consistency.csv', index=False)
```

```python
consistency.dropna(inplace=True) # remove all NaNs
consistency = consistency[consistency['fg3a'] >= 6.0] # minimum 4 attempts per game (volume shooters)
consistency = consistency[consistency['fg3_stddev']>0.01] # not sure what went wrong with my computation here
len(consistency)
```

```python
def get_image_loc(x, y, xsize, ysize):
    xstart = x-xsize/2.0
    xstop = x+xsize/2.0
    ystart = y-ysize/2.0
    ystop = y+xsize/2.0
    return xstart, xstop, ystart, ystop
```

```python
ratio = 8000
xmin, xmax = 0.2, 0.48
ymin, ymax = -0.05, -0.375
linecolor = '#000000'
titlefont = {'fontname':'Verdana',
            'fontsize': 60}
textfont = {'fontname':'Verdana',
            'fontsize': 35}
subfont = {'fontname':'Verdana',
            'fontsize': 25}
xsize, ysize = imgs[0].size
xsize /= ratio
ysize /= ratio

fig, ax = plt.subplots()
ax.annotate('       ', xy=(0.4,-0.35), xytext=(0.3,-0.35),
            arrowprops=dict(facecolor='black', shrink=1.0))
ax.annotate('       ', xy=(0.205, -0.2), xytext=(0.205, -0.3),
            arrowprops=dict(facecolor='black', shrink=1.0))
ax.text(0.21, -0.25, 'Consistency',**textfont) 
ax.text(0.335, -0.34, 'Accuracy',**textfont)
ax.text(0.29, -0.055, 'game-by-game 3p% consistency, min 6 attempts per game',**subfont)
ax.text(0.205, -0.37, 'through Nov 29 2021, created by @pkseeg',**subfont)
for x0, y0, img in zip(x, y, imgs):
    ax.imshow(img, extent=get_image_loc(x0,-y0,xsize,ysize), aspect='auto')
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymax, ymin)
ax.set_xticks([])
ax.set_yticks([])
set_size(20, 20)
ax.set_title('3-Point% Consistency Landscape',**titlefont)
plt.savefig('figures/shooting_consistency.png', bbox_inches='tight', facecolor='#FFFFFF')
plt.show()
```
