```python
# compare two teams' point differentials over the season
from basketball_reference_scraper.seasons import get_schedule
from basketball_reference_scraper.box_scores import get_box_scores

from utils import name_swap, set_size, team_colors
from tqdm.notebook import tqdm, trange
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def get_point_diffs(team, season):
    s = get_schedule(season, playoffs=False)
    t = s[(s.VISITOR == name_swap[team]) | (s.HOME == name_swap[team])]
    point_diff = {
        'game' : [],
        'point_diff': [],
        'cum_diff': []
    }
    cum = 0
    for r in trange(len(t)):
        home_score = float(t.iloc[r]['HOME_PTS'])
        away_score = float(t.iloc[r]['VISITOR_PTS'])
        diff = home_score - away_score
        if t.iloc[r]['HOME'] != name_swap[team]: diff *= -1
        cum += diff
        point_diff['game'].append(r)
        point_diff['point_diff'].append(diff)
        point_diff['cum_diff'].append(cum)
    return point_diff
```

```python
team_diffs = {}
teams = ['BOS','UTA']
for t in teams:
    team_diffs[t] = get_point_diffs(team=t, season=2022)
```

```python
dfs = []
for t in teams:
    df = pd.DataFrame(team_diffs[t])
    df.dropna(inplace=True)
    dfs.append(df)
```

```python
colors = [team_colors[t] for t in teams]
alphas = [0.4 if t=='UTA' else 0.4 for t in teams]
ymin = 10000
ymax = -10000
font = {'fontname':'Cousine',
        'fontsize': 70}
subfont = {'fontname':'Cousine',
        'fontsize': 50}

fig, ax = plt.subplots()
for df, color, alpha in zip(dfs, colors, alphas):
    ax.fill_between(df['game'],df['cum_diff'], color=color, step='pre', alpha=alpha)
    ymin = min(ymin, min(df['cum_diff']))
    ymax = max(ymax, max(df['cum_diff']))
ax.text(0,ymax,'Season Point Differentials (by game)',**font,va='top',ha='left')
ax.text(0,ymax-5,'UTA vs BOS',**font,va='top',ha='left')

# add UTA logo
uta_logo = Image.open('media/utah_logo.png')
axins0 = ax.inset_axes(bounds=(0.87, 0.955, 0.05, 0.025))
axins0.imshow(uta_logo,aspect='auto')
axins0.axis('off')

# add BOG logo
bos_logo = Image.open('media/bos_logo.png')
axins1 = ax.inset_axes(bounds=(0.915, 0.34, 0.05, 0.025))
axins1.imshow(bos_logo,aspect='auto')
axins1.axis('off')

ax.text(9.5,-3,'graph by @pkseeg stats via @py_ball',**subfont,va='top',ha='left')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
set_size(30, 70)
plt.savefig('figures/uta_bos_point_diffs.png', bbox_inches='tight', facecolor='#FFFFFF')
plt.show()
```
