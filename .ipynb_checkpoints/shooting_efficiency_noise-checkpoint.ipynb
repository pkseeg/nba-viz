```python
from basketball_reference_scraper.seasons import get_schedule
from basketball_reference_scraper.box_scores import get_box_scores
import pandas as pd
import matplotlib.pyplot as plt
from utils import set_size, name_swap, MAIN, OTHER, BACKGROUND
```

```python
def stats_by_game(season = 2021, team = 'UTA', player = 'Team Totals'):
    ''' stats_by_game
     params:
      season
      - int, season from which to pull schedule (2020-21 would be '2021')
      team
      - str, 3 character team identifier (Utah Jazz would be 'UTA')
      player
      - str, player name or 'Team Totals' for the whole team
    '''
    s = get_schedule(season, playoffs=False)
    t = s[(s.VISITOR == name_swap[team]) | (s.HOME == name_swap[team])]
    game_num = []
    # threes
    game_thp_perc = []
    cum_thp_perc = []
    thp_total = 0
    thpa_total = 0
    # field goals
    game_fg_perc = []
    cum_fg_perc = []
    fg_total = 0
    fga_total = 0
    # free throws
    game_ft_perc = []
    cum_ft_perc = []
    ft_total = 0
    fta_total = 0
    # count the missing games
    missing = []
    for r in range(len(t)):
        d = str(t.iloc[r]['DATE'])[:10]
        home = t.iloc[r]['HOME']
        visitor = t.iloc[r]['VISITOR']
        h = name_swap[home]
        v = name_swap[visitor]
        try:
            b = get_box_scores(d, v, h, period='GAME', stat_type='BASIC')[team]
            p = b[b['PLAYER']==player]
            if len(p) == 0: # this player didn't play, all are the same
                game_num.append(r)
                game_thp_perc.append(0.0)
                cum_thp_perc.append(thp_total/thpa_total)
                game_fg_perc.append(0.0)
                cum_fg_perc.append(fg_total/fga_total)
                game_ft_perc.append(0.0)
                cum_ft_perc.append(ft_total/fta_total)
                continue
            # game number
            game_num.append(r)
            
            # threes
            thp = float(p['3P'])
            thpa = float(p['3PA'])
            #   game
            if thpa == 0: game_thp_perc.append(0.0)
            else: game_thp_perc.append(thp/thpa)
            #   cumulative
            thp_total += thp
            thpa_total += thpa
            if thpa_total == 0: cum_thp_perc.append(0.0)
            else: cum_thp_perc.append(thp_total/thpa_total)
            
            # fg
            fg = float(p['FG'])
            fga = float(p['FGA'])
            #   game
            if fga == 0: game_fg_perc.append(0.0)
            else: game_fg_perc.append(fg/fga)
            #   cumulative
            fg_total += fg
            fga_total += fga
            if fga_total == 0: cum_fg_perc.append(0.0) 
            else: cum_fg_perc.append(fg_total/fga_total) 
            
            
            # ft
            ft = float(p['FT'])
            fta = float(p['FTA'])
            #   game
            if fta == 0: game_ft_perc.append(0.0)
            else: game_ft_perc.append(ft/fta)
            #   cumulative
            ft_total += ft
            fta_total += fta
            if fta_total == 0: cum_ft_perc.append(0.0) 
            else: cum_ft_perc.append(ft_total/fta_total)
        except BaseException as e:
            print(d, h, v)
            print(e)
            missing.append((d, h, v))
    df = pd.DataFrame()
    df['game_num'] = game_num
    df['game_3p'] = game_thp_perc
    df['cum_3p'] = cum_thp_perc
    df['game_fg'] = game_fg_perc
    df['cum_fg'] = cum_fg_perc
    df['game_ft'] = game_ft_perc
    df['cum_ft'] = cum_ft_perc
    print('Missing: '+str(missing))
    return df
```

```python
don = stats_by_game(season=2021,team='UTA',player='Donovan Mitchell') # do donovan's first
```

```python
don = don.head(72) # remove all the playoff games (shouldn't be including them but it is...)
```

```python
jazz = stats_by_game(season=2021,team='UTA') # jazz as a team
```

```python
jazz = jazz.head(72) # again, we need to remove playoffs
```

```python
fig, axs = plt.subplots(3, figsize=(10,20))
fig.tight_layout(pad=0.0)
# plot one
last_val = don.iloc[-1]['cum_3p']
axs[0].plot(don['game_num'], don['cum_3p'], lw=2, color = MAIN)
axs[0].plot([0, 72], [last_val, last_val], '--', lw=1, color = MAIN)
axs[0].text(20,last_val-last_val*0.02,'Don: '+str(round(last_val,3)), 
         color = MAIN,
         fontsize = 'x-large')
last_val = jazz.iloc[-1]['cum_3p']
axs[0].plot(jazz['game_num'], jazz['cum_3p'], lw=2, color = OTHER)
axs[0].plot([0, 72], [last_val, last_val], '--', lw=1, color = OTHER)
axs[0].text(1,last_val+last_val*0.015,'Jazz: '+str(round(last_val,3)), 
         color = OTHER,
         fontsize = 'x-large')

axs[0].set_title('Team vs Individual Cumulative Shooting Efficiency 2020-21')
axs[0].set_ylabel('Season 3P%')
axs[0].set_ylim(0.25, 0.43)
axs[0].set_xticks([])
# plot two
last_val = don.iloc[-1]['cum_fg']
axs[1].plot(don['game_num'], don['cum_fg'], lw=2, color = MAIN)
axs[1].plot([0, 72], [last_val, last_val], '--', lw=1, color = MAIN)
axs[1].text(25,last_val+last_val*0.015,'Don: '+str(round(last_val,3)), 
         color = MAIN,
         fontsize = 'x-large')
last_val = jazz.iloc[-1]['cum_fg']
axs[1].plot(jazz['game_num'], jazz['cum_fg'], lw=2, color = OTHER)
axs[1].plot([0, 72], [last_val, last_val], '--', lw=1, color = OTHER)
axs[1].text(1,last_val+last_val*0.015,'Jazz: '+str(round(last_val,3)), 
         color = OTHER,
         fontsize = 'x-large')
axs[1].set_ylabel('Season FG%')
axs[1].set_ylim(0.30, 0.50)
axs[1].set_xticks([])
# plot three
last_val = don.iloc[-1]['cum_ft']
axs[2].plot(don['game_num'], don['cum_ft'], lw=2, color = MAIN)
axs[2].plot([0, 72], [last_val, last_val], '--', lw=1, color = MAIN)
axs[2].text(25,last_val+last_val*0.015,'Don: '+str(round(last_val,3)), 
         color = MAIN,
         fontsize = 'x-large')
last_val = jazz.iloc[-1]['cum_ft']
axs[2].plot(jazz['game_num'], jazz['cum_ft'], lw=2, color = OTHER)
axs[2].plot([0, 72], [last_val, last_val], '--', lw=1, color = OTHER)
axs[2].text(2,last_val+last_val*0.015,'Jazz: '+str(round(last_val,3)), 
         color = OTHER,
         fontsize = 'x-large')
axs[2].set_ylabel('Season FT%')
axs[2].set_xlabel('Game Number')
axs[2].set_ylim(0.65, 1.0)

plt.savefig('figures/shooting_efficiency_noise.png', bbox_inches='tight', facecolor='#FFFFFF')
plt.show()
```
