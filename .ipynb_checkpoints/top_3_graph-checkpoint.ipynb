```python
# Warriors, Suns, Jazz ratings
from basketball_reference_scraper.seasons import get_schedule
from basketball_reference_scraper.box_scores import get_box_scores

from utils import name_swap, set_size
from tqdm.notebook import tqdm, trange
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_point_diffs(team, season):
    s = get_schedule(season, playoffs=False)
    t = s[(s.VISITOR == name_swap[team]) | (s.HOME == name_swap[team])]
    #print(t)
    point_diff = {
        'game' : [],
        'point_diff': []
    }
    for r in trange(len(t)):
        home_score = float(t.iloc[r]['HOME_PTS'])
        #print(home_score)
        away_score = float(t.iloc[r]['VISITOR_PTS'])
        #print(away_score)
        diff = home_score - away_score
        #print(t.iloc[r]['HOME'])
        #print(name_swap[team])
        if t.iloc[r]['HOME'] != name_swap[team]: diff *= -1
        #print(diff)
        point_diff['game'].append(r)
        point_diff['point_diff'].append(diff)
    return point_diff
```

```python
team_diffs = {}
for t in ['PHO', 'GSW', 'UTA']:
    team_diffs[t] = get_point_diffs(team=t, season=2022)
```

```python
uta_df = pd.DataFrame(team_diffs['UTA'])
pho_df = pd.DataFrame(team_diffs['PHO'])
gsw_df = pd.DataFrame(team_diffs['GSW'])
dfs = [uta_df, pho_df, gsw_df]
```

```python
colors = ['#000000', '#1D1160', '#FFC72C']
alphas = [1.0, 0.5, 0.5]
width = 0.25#]*len(df['game'])
p = 0

fig, ax = plt.subplots()
for df, color, alpha in zip(dfs, colors, alphas):
    ax.bar(df['game'] + p, df['point_diff'], color=color, alpha=alpha, width=width)
    #if alpha == 0.2: break
    p += width
set_size(30, 70)
plt.show()
```
