```python
from py_ball.team import Team
from py_ball.image import Logo

# data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# headers
from config import py_ball_headers

# tqdm
from tqdm.notebook import tqdm

# team ids
from utils import team_ids, ID_TO_TEAM_NBA, set_size, team_colors
```

```python
MAIN = 'UTA'
OTHER = 'BOS'
two_teams = [MAIN, OTHER]
team_info = {
  'TEAM_ID': [],
  'GP': [],
  'W': [],
  'L': [],
  'W_PCT': [],
  'MIN': [],
  'E_OFF_RATING': [],
  'OFF_RATING': [],
  'E_DEF_RATING': [],
  'DEF_RATING': [],
  'E_NET_RATING': [],
  'NET_RATING': [],
  'AST_PCT': [],
  'AST_TO': [],
  'AST_RATIO': [],
  'OREB_PCT': [],
  'DREB_PCT': [],
  'REB_PCT': [],
  'TM_TOV_PCT': [],
  'EFG_PCT': [],
  'TS_PCT': [],
  'E_PACE': [],
  'PACE': [],
  'PACE_PER40': [],
  'POSS': [],
  'PIE': []
}
```

```python
for t in tqdm(two_teams):
    t_id = team_ids[t]
    team = Team(py_ball_headers,
                team_id=t_id,
                per_mode='Per100Plays',
                measure_type='Advanced',
                season='2021-22')
    team = team.data['OverallTeamDashboard'][0]
    for k in team_info:
        if k == 'TEAM_ID':
            team_info[k].append(t_id)
        else:
            team_info[k].append(team[k])
```

```python
df = pd.DataFrame(team_info)
df['TEAM_ABBREVIATION'] = [ID_TO_TEAM_NBA[i] for i in df['TEAM_ID']]
```

```python
df.head()
```

```python
key_factors = ['OFF_RATING', 'DEF_RATING', 'NET_RATING', 
               'AST_RATIO', 'REB_PCT', 'OREB_PCT', 'DREB_PCT',
               'TM_TOV_PCT', 'EFG_PCT', 'TS_PCT', 
               'PACE', 'TEAM_ABBREVIATION']
df = df[key_factors]
```

```python
main_row = df[df['TEAM_ABBREVIATION']==MAIN]
other_row = df[df['TEAM_ABBREVIATION']==OTHER]
```

```python
# params
colors = [team_colors[t] for t in two_teams]
alpha_top = 1.0
alpha_bottom = 0.5
width = 1.0
top_diff = 100
bottom_diff = 100
ann = {'fontname':'Verdana',
       'fontsize': 70}
subann = {'fontname':'Verdana',
          'fontsize': 40}

# plots
# 1. offense vs defense [X]
#  FIXME add lines for league average ?
# 2. assist ratio, team turnover percentage [ ]
# 3. rebounding, pace [ ]
# 4. shooting [ ]
fig, axs = plt.subplots(nrows=2,ncols=2)

# 1.
off_poss = [main_row['OFF_RATING'].values[0]-top_diff,other_row['DEF_RATING'].values[0]-top_diff]
def_poss = [-main_row['DEF_RATING'].values[0]+bottom_diff,-other_row['OFF_RATING'].values[0]+bottom_diff]
ymin, ymax = -9, 16
xmin, xmax = -1, 7
axs[0,0].bar([0,1], off_poss, color=colors, alpha=alpha_top, width=width)  
axs[0,0].bar([0,1], def_poss, color=colors, alpha=alpha_bottom, width=width)
axs[0,0].text((xmax-xmin)//2-2, ymax//2, MAIN+' Offensive Possesions', **ann)
axs[0,0].text((xmax-xmin)//2-2, ymax//2-1, MAIN+' Off. Rating vs '+OTHER+' Def. Rating', **subann)
axs[0,0].text((xmax-xmin)//2-2, ymin//2, MAIN+' Defensive Possesions', **ann)
axs[0,0].text((xmax-xmin)//2-2, ymin//2-1, MAIN+' Def. Rating vs '+OTHER+' Off. Rating', **subann)
axs[0,0].set_ylim(ymin, ymax)
axs[0,0].set_xlim(xmin, xmax)
axs[0,0].spines['top'].set_visible(False)
axs[0,0].spines['right'].set_visible(False)
axs[0,0].spines['bottom'].set_visible(False)
axs[0,0].spines['left'].set_visible(False)
axs[0,0].get_xaxis().set_ticks([])
axs[0,0].get_yaxis().set_ticks([])
set_size(40,80)
plt.show()
```
