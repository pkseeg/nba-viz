```python
# rudy-pallete
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
from utils import MAIN, OTHER, BACKGROUND

# look at fonts and stuff
#plt.rcParams.keys()
font = {'font.monospace' : 'Courier New',
        'font.size'   : 11}
plt.rcParams.update(font)
rudy_patch = mpatches.Patch(color=MAIN, label='rudy gobert')
other_patch = mpatches.Patch(color=OTHER, label='anyone else')
```

```python
# DATA
from bbref_scraper.player_season_advanced import PlayerSeasonAdvanced

# player_season_per_game for 2022 (current season)
psa = PlayerSeasonAdvanced(2022)

psa.df = psa.df.drop(columns=['DUMMY'])
psa.df.columns
```

```python
psa.df['dws'] = pd.to_numeric(psa.df['dws'], downcast='float')
psa.df['ows'] = pd.to_numeric(psa.df['ows'], downcast='float')
psa.df['dbpm'] = pd.to_numeric(psa.df['dbpm'], downcast='float')
```

```python
top10_dws = psa.df.sort_values(by='dws', ascending=False).head(10)
colors = []
for p in top10_dws['player']:
    colors.append(MAIN if p == 'Rudy Gobert' else OTHER)
ax = sns.barplot(x='player', y='dws', data=top10_dws, palette = colors)
ax.set_facecolor(BACKGROUND)
ax.set(title = 'defensive win shares so far 2021-22', 
       xlabel='', 
       ylabel='defensive win shares')
ax.set(ylim=(0, 0.7))
plt.xticks(rotation=-90, fontsize = 'small')
#plt.tick_params(axis='x',          
#                which='both',      
#                bottom=False,      
#                top=False,        
#                labelbottom=False)
plt.legend(handles=[rudy_patch, other_patch])
plt.show()
```

```python
top10_ows = psa.df.sort_values(by='ows', ascending=False).head(10)
colors = []
for p in top10_ows['player']:
    colors.append(MAIN if p == 'Rudy Gobert' else OTHER)
ax = sns.barplot(x='player', y='ows', data=top10_ows, palette = colors)
ax.set_facecolor(BACKGROUND)
ax.set(title = 'offensive win shares so far 2021-22', 
       xlabel='', 
       ylabel='offensive win shares')
ax.set(ylim=(0, 1.2))
plt.xticks(rotation=-90, fontsize = 'small')
#plt.tick_params(axis='x',          
#                which='both',      
#                bottom=False,      
#                top=False,        
#                labelbottom=False)
plt.legend(handles=[rudy_patch, other_patch])
plt.show()
```

```python
top10_dws = psa.df.sort_values(by='drb_pct', ascending=False).head(10)
colors = []
for p in top10_dws['player']:
    colors.append(MAIN if p == 'Rudy Gobert' else OTHER)
ax = sns.barplot(x='player', y='drb_pct', data=top10_dws, palette = colors)
ax.set_facecolor(BACKGROUND)
ax.set(title = 'defensive win shares so far 2021-22', 
       xlabel='', 
       ylabel='defensive win shares')
ax.set(ylim=(0, 100))
plt.xticks(rotation=-90, fontsize = 'small')
#plt.tick_params(axis='x',          
#                which='both',      
#                bottom=False,      
#                top=False,        
#                labelbottom=False)
plt.legend(handles=[rudy_patch, other_patch])
plt.show()
```

```python
colors = []
for p in psa.df['player']:
    colors.append(MAIN if p == 'Rudy Gobert' else OTHER)
psa.df['sizes'] = [60.0 if p == 'Rudy Gobert' else 10.0 for p in psa.df['player']]
```

```python
ax = sns.scatterplot(x='dws', 
                     y='ows', 
                     data=psa.df, 
                     hue='player', 
                     size ='sizes',
                     palette = colors,
                     alpha=.75)
ax.set_facecolor(BACKGROUND)
ax.set(title = 'win shares so far 2021-22', 
       xlabel='defensive win shares', 
       ylabel='offensive win shares')
ax.set(ylim=(0, 1.5))
plt.legend(handles=[rudy_patch, other_patch])
plt.show()
```
